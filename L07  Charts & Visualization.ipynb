```python
total = Table.read_table('SchoolEnrollmentPune.csv')
total = total.select('SchName', 'Tot G_0-12', 'Tot B_0-12', 'total', 'teachers')
total
```

# LINE PLOT

```python
#THIS EXAMPLE OF THE LINE PLOT IS NOT QUITE USEFUL FOR INTERPRETATION.
#USE SCATTER PLOT INSTEAD

#Use line plots for sequential quantitative data: if...
#... x-axis has an order
#... sequential in differences in y values are meaningful
#=================#... only one y value for each x-value ======================
#...Often x-axis is time or distance

#Use scatter plots for non-sequential quantitative data
```

```python
total.plot('total', 'teachers')
```

```python
#Take a close look.
#In The second line, total1000 = total1000.where

total1000 = total.where('Tot G_0-12', are.above_or_equal_to(1000))
total1000 = total1000.where('Tot B_0-12', are.above_or_equal_to(1000))
total1000
```

```python
total1000.plot('total', 'teachers')
plots.xlim(2000, 6000); #Not needed for Data8
```

# SCATTER PLOT


```python
totalkids_arr = total.column('Tot G_0-12') + total.column('Tot B_0-12')
totalkids_arr
```

```python
total = total.with_column('Total', totalkids_arr)
teacher_per_student = total.column('Total') / total.column('teachers')
total = total.with_column('Students per Teacher', teacher_per_student)
total
```

```python
#t.scatter(x_label, y_label)

total.scatter('Total', 'teachers')
```

```python
total.scatter('Students per Teacher', 'Total')
```

```python
#Notice
#We have two schools with more than 6000 students that have around 
#40 students in a class

#Assuming we let Students per Teacher = Class Strength


#Also, there are many schools with less than a thousand students where the
#class strength is high
```
